Machine Learning

```python
# Divide the dataframe
X = df.drop("Class", axis=1)
y = df.Class
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, stratify=y, random_state=42)
X_train
```

```python
# Instaniate the models
scaler = StandardScaler()
log_reg = LogisticRegression()
```

```python
# Scale the Features and fit the model
X_train_scaled = scaler.fit_transform(X_train)
log_reg.fit(X_train_scaled, y_train)
```

```python
X_test
```

```python
# Evaluate the model on the test set
X_test_scaled = scaler.transform(X_test)
# The reason we make use transform not fit_transform is to avoid data leakage. if the scaler is fitted to the test data, it will falsify the results of our evaluation. And that certainly isn't what we want.
y_pred = log_reg.predict(X_test_scaled)
print(confusion_matrix(y_test, y_pred))
sns.heatmap(confusion_matrix(y_test, y_pred), cmap='coolwarm', annot=True)
```

```python
# Evaluate the model on the test set using 4 different metrics
print(f"Accuracy Score: {accuracy_score(y_test, y_pred):.4f}")
print(f"Precision Score: {precision_score(y_test, y_pred):.4f}")
print(f"Recall Score: {recall_score(y_test, y_pred):.4f}")
print(f"F1 Score: {f1_score(y_test, y_pred):.4f}")
```

```python
# Evaluate the model on the training set
y_pred_train = log_reg.predict(X_train_scaled)
print(f"Accuracy Score: {accuracy_score(y_train, y_pred_train):.4f}")
print(f"Precision Score: {precision_score(y_test, y_pred):.4f}")
print(f"Recall Score: {recall_score(y_test, y_pred):.4f}")
print(f"F1 Score: {f1_score(y_test, y_pred):.4f}")
```
